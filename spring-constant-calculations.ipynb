```python
import numpy as np
```

A very simple spring constant calculation table

```python
D = np.array([4.5, 5, 4, 5.5])
L = np.array([44.5, 20.5, 79.5, 25.5])
d = 7.346685794e-4 # calculated, this in metres
n = L / d # an appropriate approximation

#n = float(input("n: "))
#D = float(input("D: "))
D = D*10**-3
(G_lower, G_higher) = np.array([77e9, 81e9]) # this in the lower bound (shoudl include some actual range)


k_lower = (G_lower * (d)**4) / (8 * (D)**3 * n)
k_higher = (G_higher * (d)**4) / (8 * (D)**3 * n)
k = np.array([np.mean([k_lower[i], k_higher[i]]) for i in range(len(k_lower))])

print(k)
```

Now I have the $k$ values for each of the springs (these feel incredibly small so these may just be wrong), I will try to compute the necessary forces for the given extension limits. I hope they are small.

```python
# f = k * \Delta x
dx = 5e-3

F = k * dx
print(F)
```
